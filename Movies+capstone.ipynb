# Kaggle Film
This is a film dataset from Kaggle. 
This data set contains 46,907 films the year they were released, genre, duration, country, actors, average vote score and number of votes. three questiosns from this data that I would like to know is

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
year=moviedf.sort_values('year')
year1=year.groupby('year').count()
yearbins=year['year'].unique()
yearbins

```

```python
year1
```

```python
plt.bar(yearbins,year1['film_title'],fill=False)
plt.xlabel('Year of Films Release')
plt.ylabel('How many movies in year')
plt.title('Films made since 1911 by year')
plt.show()
```

From the chart, films are being created at a much higher rate in later decades. 
It shows an upward rising trend gradually rising until the alte 70's where it takes a slight dip
before makeing a large jump in the late 80's early 90's 
There is a large spike in mid 00'S, possibly from film making turing digital 



```python
moviesortedgenre = moviedf.sort_values('genre')
moviesortedgenre=moviesortedgenre.dropna()
moviesortedgenre.tail()
```

```python
moviesortedgenre['genre'].unique()
```

```python
movieavgs=moviesortedgenre.groupby('genre').aggregate(np.mean)
movieavgs
```

I have noticed that the average vote amount is skewed for some genres. For this reason, I am removing any films that have less than 14 votes in total. The reason that 14 was selected because the data showed that the genres with less than 14 votes avg ratings were amoungst the highest. 

```python
movieavg14=movieavgs.votes>14
movie14=movieavgs[movieavg14]
movie14
```

```python
heights = [10, 20, 15]
bars = ['Action', 'Adventure','Animation','Biblical','Biography','Comedy','Crime','Drama','Fantasy','Gangster','Grotesque','History','Horror','Musical','Mythology','Mélo','Noir','Sport','Spy','Super-hero','Thriller', 'War','Western']
y_pos = range(len(bars))
plt.bar(y_pos, movie14['avg_vote'])
# Rotation of the bars names
plt.xlabel('Movie Genres')
plt.ylabel('Average Rating')
plt.xticks(y_pos, bars, rotation=90)
plt.show()
```

```python
noi= moviedf['genre']=='Noir'
noir=moviedf[noi]
gang=moviedf['genre']=='Gangster'
gangster=moviedf[gang]
gro=moviedf['genre']=='Grotesque'
grotesque=moviedf[gro]
wes= moviedf['genre']=='Western'
western=moviedf[wes]
hor=moviedf['genre']=='Horror'
horror=moviedf[hor]
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(noir['year'], noir['avg_vote'],'o')
plt.xlabel('Years')
plt.ylabel('Avg vote')
plt.title('Film Noir average vote distrabution')
plt.subplot(1,2,2)
plt.plot(noir['year'],noir['duration'],'o')
plt.xlabel('Years')
plt.ylabel('Duration in min')
plt.title('Film Noir duration distrabution')

plt.show()
```

Film Noir began in the 1920's but recieved alot of high votes during the 40's when it became very popular. Possibly due to World War II. People would watch movies to keep their spirits High.
As the lengths of the films rise votes lower. But this could also be because consumer begain to watch more color films as time goes on. Would need further reasearch into this.
